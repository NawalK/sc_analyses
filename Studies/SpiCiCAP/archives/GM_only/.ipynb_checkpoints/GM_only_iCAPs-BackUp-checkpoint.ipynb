```python
from os import sys
sys.path.append('/media/miplab-nas2/Data3/Nawal/')
import ConsensusClustering
import numpy as np
import nibabel as nb
import matplotlib.pyplot as plt
import seaborn as sns
import pandas as pd
from scipy.io import loadmat
from numpy import median
```

### Parameters

```python
Metadata = {
    'Root': '/media/miplab-nas2/Data/SpinalCord/3_RestingState/LongRecordings/Cervical/iCAPs_results/',
    'icapFolder': 'PAM50_light_gm_19sub_Alpha_5_950DOT05/',
    'consFolder': '5to20_SubsampleType_items_Fraction_0DOT8_nFolds_10_Dist_cosine',
    'K': [*range(5,21)],
            }
```

### Define number of K (consensus clustering)

**1. Area under the curve**

```python
AUC = loadmat(Metadata['Root']+Metadata['icapFolder']+Metadata['consFolder']+'/AUC.mat')['AUC']
deltaAUC = np.diff(AUC,axis=0)
```

**2. Average consensus**

```python
#cons_k = {cons: [] for cons in Metadata['K']}
consval = np.empty((0,), float)
kval = np.empty((0,), int)
for k in Metadata['K']:
    #cons_k[k]=loadmat(Metadata['Root']+Metadata['icapFolder']+'K_'+str(k)+'_Dist_cosine_Folds_20/iCAPs_consensus.mat')['iCAPs_consensus']
    #consval.append(np.squeeze(loadmat(Metadata['Root']+Metadata['icapFolder']+'K_'+str(k)+'_Dist_cosine_Folds_20/iCAPs_consensus.mat')['iCAPs_consensus']))
    consval = np.append(consval,loadmat(Metadata['Root']+Metadata['icapFolder']+'K_'+str(k)+'_Dist_cosine_Folds_20/iCAPs_consensus.mat')['iCAPs_consensus'])
    kval = np.append(kval,(np.full((k,),k)))
    
colnames = ["k","consensus"]
cons_df = pd.DataFrame(list(zip(kval,consval)),columns=colnames)
```

**3. Plot both**

```python
fig, axes = plt.subplots(1, 2,figsize=(15,4))
fig.suptitle('Select K values (consensus clustering)',fontweight="bold",fontsize=13)
axes[0].set_title(r'$\Delta$AUC'); axes[1].set_title('Average consensus',fontsize=13);
auc=sns.lineplot(data=deltaAUC,ax=axes[0],markers=True,palette="flare");
auc.set_xticks(range(0,15))
auc.set_xticklabels(['5-6','6-7','7-8','8-9','9-10','10-11','11-12','12-13','13-14','14-15','15-16','16-17','17-18','18-19','19-20'], rotation = 90)
sns.pointplot(data=cons_df,x="k", y="consensus",ax=axes[1],palette="flare",ci='sd',err_style='bars');
```

**4. Look at CDF as well**

```python
CDF = loadmat(Metadata['Root']+Metadata['icapFolder']+Metadata['consFolder']+'/CDF.mat')['CDF']
```

```python
kval = np.empty((0,), int) # K range
consval = np.empty((0,), int) # From 0 to 100 for each K
cdfval = np.empty((0,), float) # Fraction < consensus value
for k_ix, k in enumerate(Metadata['K']):
    for cons_ix in range(0,len(CDF[k_ix,])):
        kval = np.append(kval,k)
        consval = np.append(consval,cons_ix)
        cdfval = np.append(cdfval,CDF[k_ix,cons_ix])
        
colnames = ["k","consensus","fraction"]
cdf_df = pd.DataFrame(list(zip(kval,consval,cdfval)),columns=colnames)
```

```python
# Style palette
dash_list = sns._core.unique_dashes(cdf_df['k'].unique().size+1)
style = {key:value for key,value in zip(cdf_df['k'].unique(), dash_list[1:])}
style[6] = ''  # empty string means solid
style[9] = ''
style[11] = ''
style[14] = ''
# Size palette
linesize = {key:value for key,value in zip(cdf_df['k'].unique(), np.full((len(Metadata['K']),),1))}
linesize[6] = 2  # empty string means solid
linesize[9] = 2
linesize[11] = 2
linesize[14] = 2
```

```python
sns.set_style("ticks")
plt.figure(figsize=(7,6))
sns.lineplot(data=cdf_df,x='consensus',y='fraction',hue='k',style='k',size='k',sizes=linesize,dashes=style,palette=sns.color_palette('flare', n_colors=16));
plt.legend(bbox_to_anchor=(1.01,1.02), loc='upper left',frameon=False);
plt.tight_layout()
```
